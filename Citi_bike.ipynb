```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
jan_df = pd.read_csv(Path("JC-202301-citibike-tripdata.csv"))
jan_df.dropna(how='any', inplace=True)
jan_df.info()
jan_df.head()
```

```python
feb_df = pd.read_csv(Path("JC-202302-citibike-tripdata.csv"))
feb_df.dropna(how='any', inplace=True)
feb_df.info()
feb_df.head()
```

```python
mar_df = pd.read_csv(Path("JC-202303-citibike-tripdata.csv"))
mar_df.dropna(how='any', inplace=True)
mar_df.info()
mar_df.head()
```

```python
apr_df = pd.read_csv(Path("JC-202304-citibike-tripdata.csv"))
apr_df.dropna(how='any', inplace=True)
apr_df.info()
apr_df.head()
```

```python
may_df = pd.read_csv(Path("JC-202305-citibike-tripdata.csv"))
may_df.dropna(how='any', inplace=True)
may_df.info()
may_df.head()
```

```python
jun_df = pd.read_csv(Path("JC-202306-citibike-tripdata.csv"))
jun_df.dropna(how='any', inplace=True)
jun_df.info()
jun_df.head()
```

```python
july_df = pd.read_csv(Path("JC-202307-citibike-tripdata.csv"))
july_df.dropna(how='any', inplace=True)
july_df.info()
july_df.head()
```

```python
citi_bike_df = jan_df.append(feb_df).append(mar_df).append(apr_df).append(may_df).append(jun_df).append(july_df)
citi_bike_df.info()
citi_bike_df.head()
```

```python
citi_bike_df.to_csv("JC-Citi-Bike.csv")
```
